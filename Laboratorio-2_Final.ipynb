## 3. Soldados Enfermos

En el ejército de Sansalandia se propagó un virus cuyos síntomas no son percibidos por el huésped hasta que es demasiado tarde. Gracias a la avanzada tecnología se ha logrado diseñar un test que puede determinar si el individuo está infectado (positivo) o no (negativo) utilizando una muestra de sangre del individuo, pero tiene un costo alto el hecho de realizarlo por individuo $C$. 

El ejército ideó un plan para intentar reducir los costos y consiste en agrupar los $N$ soldados. Se dividen en $M$ grupos de $K$ soldados de forma que $K\cdot M = N$ y se agrupan las $K$ muestras de sangre en una sola para realizar el test. Si uno de cada diez soldados es afectado por el virus responda las siguiente preguntas: 

* $C = 500$

>a) Compare el costo promedio utilizando la agrupación $K$ y el costo al aplicar el test a cada soldado uno por uno. Asuma que si un grupo arroja positivo se hace el test uno por uno (de los pertenecientes al grupo).

```r
#Valores iniciales
C = 500
N = 3000
K = 60
Prob = 0.1
M = N/K
#Costo acumulado
S_costo = 0
simulaciones = 10000
#Se comienza con las tomando las K muestras de sangre:
for (i in 1:simulaciones){
    costo = 0
    #Se hace las prueba en los M grupos
    for (i in 1:M){
        randint <- sample(1:10, 1)
        if (randint == 1){
            costo = costo + (C*(K-1))
        }
        else{
        costo = costo + C
        }  
    }
    S_costo = S_costo + costo
}
#Se realiza un costo promedio
costo_prom = S_costo/simulaciones
print("El Costo promedio calculado es:")
print(costo_prom)
costo_total = N*C
print("El Costo total es:")
print(costo_total)
```

#### Respuesta

Queremos saber la probabilidad de que algún grupo tenga un resultado positivo, para esto observamos que los soldados se dividen en M grupos y que el virus esta en alguna persona representante del grupo i $\in$ {0,1,2,...,M}. Para plantear un modelo que se ajuste a la realidad, tendremos un suceso $A_i$, que es encontrar una persona enferma en el grupo i. Tambien asumimos que la enfermedad no es contagiosa por lo que cada caso de enfermedad es independiente uno del otro. Una vez dicho esto tenemos la siguiente planteo:  
P =  
P($A_1$ ∩ ~$A_2$ ∩ ~$A_3$ ∩ ... ∩ ~$A_M$) U  
P(~$A_1$ ∩ $A_2$ ∩ ~$A_3$ ∩ ... ∩ ~$A_M$) U

P(~$A_1$ ∩ ~$A_2$ ∩ ~$A_3$ ∩ ... ∩ $A_M$)  
Utilizando el método de multiples eventos independientes de la probabilidad condicional obtenemos que:  
P =
P($A_1$)$\cdot$P(~$A_2$)$\cdot$P(~$A_3$)$\cdot$...P(~$A_M$) +

P(~$A_1$)$\cdot$P(~$A_2$)$\cdot$P(~$A_3$)$\cdot$... P($A_M$)  
Como solo hay dos posibilidades, P($A_n$) = 0.1 y P(~$A_n$) = 0.9 entonces decimos que:

$M \cdot (0.1) \cdot (0.9)^{(M-1)}$  (1)  

Este fenomeno es comparable a la distribución binomial

#### Respuesta  
La Función de probabilidad de la distribución binomial es de la siguiente forma:

Bin(n,p) = $\frac{n!}{x!(n-x)!} \cdot $ $ p^{x} \cdot (1-p)^{(n-x)}$

Instanciada con x=1 y p = 0,1 => X ~ Bin(1,0.1):  

Bin(1,0.1) = $\frac{M!}{1!(M-1)!} \cdot $ $ 0.1^{1} \cdot (1-0.1)^{(M-1)}$

Una vez obtenido el resultado, se obtiene el valor esperado que para una distribución binomial es $n \cdot P$

Se procede a calcular el valor esperado:

```r
#Constantes del problema
C = 500
N = 3000
K = 60
Prob = 0.1
M = N/K
#calculo Bin(1,0.1)
P_total = M*(Prob)*((1-Prob)^(M-1))
#calculo valor esperado
valor_esperado = P_total * M
#Print valor esperado
print("valor esperado")
print(valor_esperado)
#Se estima el costo a través del valor esperado sin aproximar sumando los K integrantes del grupo analizado
#y las M revisiones a los grupos
valor_sin_aproximar = (valor_esperado*(K)+(M-1))*C
print("costo sin aproximar al valor esperado")
print(valor_sin_aproximar)
#Se estima el costo a través del valor esperado aproximado sumando los K integrantes del grupo analizado
#y las M revisiones a los grupos
valor_aproximado = (K+M-1)*C
print("costo que se aproxima al valor esperado")
print(valor_aproximado)
```
